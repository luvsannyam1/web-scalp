```python
def get_data(last):
  url = "https://portal.laekh.de/pages/rest/umkreissuche/search?type="+type+"&standort=&standort_lat=&standort_lon=&radius="+radius+"&suchfilter="+suchfilter+"&last_item_cnt="+last+"&sortfilter=0&sortdir=asc"
  payload={}
  headers = {
    'Cookie': 'JSESSIONID=D77506604CAE501EB6B7B67E3C324530'
  }
  response = requests.request("GET", url, headers=headers, data=payload)
  json_data = json.loads(response.text)
  soup = BeautifulSoup(json_data['adresslist'],'html.parser')
  print("called")
  iteration = []
  for list in soup.find_all("tr"):
    # print(list.find('div',class_='css_bold').text) # ArtzName
    
    # print(list.find('div',class_='wordwrap').text) # Ort
    # print(list.find('div',class_='padding_top wordwrap').text) # Address
    address= list.find('div',class_='padding_top wordwrap').text.split()
    # print(list.find('div',class_='padding_top css_bold').text) # department
    department = list.find('div',class_='padding_top css_bold').text.split()
    # print(department)
    # print(list.find('div',class_='',text=True)) # Duration
    # print("=============================================") # Duration
    duration = ""
    if hasattr(list.find('div',class_='',text=True),'text') is True:
      duration = list.find('div',class_='').text

    dict = {
      'subject':"",
      'name':list.find('div',class_='css_bold').text,
      'authority_since':"",
      'WBO':department[len(department)-1].replace(')',''),
      'duration' : duration,
      'address':list.find('div',class_='padding_top wordwrap').text,
      'zip':address[0],
      'city':address[1],
      'address_title':'',
      'clinic_type':list.find('div',class_='padding_top css_bold').text,
      'import_date':datetime.datetime.now().strftime("%Y-%m-%d %H-%M-%S"),
      'import_link':url
    }
    print(dict)
    iteration.append(dict)
    
  return iteration
```

```python
f = open("collected.json", "a")
collected = []
while int(last_item_cnt) < 5794:
  last_item_cnt =  str(int(last_item_cnt) + 20)
  collected.append(get_data(last_item_cnt))
f.write(json.dumps(collected,indent=2))
f.close()
```
