```python
import requests
from bs4 import BeautifulSoup
import datetime
import json
```

```python
limit = 152
current = 0
```

```python
def get_data(current):
  url = "https://secure.blaek.de/Portal/WBBSuchePortal/WBBSuchePortal/SucheEinfach"

  payload={'SuchArt': 'einfach',
  'Entfernung': '50',
  'Anzahl': '100',
  'Search': 'true',
  'Befugnis': '',
  'Arzt': '',
  'wohnort': '',
  'paginationIndex': str(current)}
  headers = {
    'Content-Type': 'application/x-www-form-urlencoded',
    'Cookie': 'ASP.NET_SessionId=s1nfztxyfh0fff4v2ayd01ih'
  }

  response = requests.request("POST", url, headers=headers, data=payload)

  soup =  BeautifulSoup(response.text,'html.parser').find('table',style="width:100%; padding-right: 14px;")
  collected = []
  for list in soup.find_all('tr'):
    dict = {
      'subject':"",
      'name':"",
      'authority_since':"",
      'WBO': "",
      'duration' : "",
      'address': "",
      'zip': "",
      'city': "",
      'address_title':'',
      'clinic_type':"",
      'import_date':datetime.datetime.now().strftime("%Y-%m-%d %H-%M-%S"),
      'import_link':url
    }
    for index , item in enumerate(list.find_all('td'),start = 0):
      item = item.text.replace('\r','').replace('\n','').replace('  ','').split(',')
      if index == 0 :
        dict['address'] = str(item[0])
      if index == 1 :
        dict['clinic_type'] = item[0]
      if index == 2 :
        dict['clinic_type'] : item[0]
    collected.append(dict)
  return collected
```

```python
f = open("collected.json", "a")
collected = []
while current <= limit:
  collected.append(get_data(current))
  current+=1
f.write(json.dumps(collected,indent=2))
f.close()
```
