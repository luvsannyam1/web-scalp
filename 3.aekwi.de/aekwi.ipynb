```python
import requests
import json
from bs4 import BeautifulSoup
import datetime
```

```python
current = 1
limit = 59
```

```python
def get_data(pagination):
  url = "https://www.aekwl.de/fuer-aerzte/weiterbildung/befugnisse-zur-weiterbildung/datenbank-weiterbildungsbefugte-aerzte/?tx_wfqbe_pi1%5Bsubmit%5D=Suchen&tx_wfqbe_pi1%5Bshowpage%5D%5B189%5D="+str(pagination)

  payload={}
  headers = {
    'Cookie': 'PHPSESSID=7bso6lb99pa7luim70n7qqu3f2'
  }

  response = requests.request("GET", url, headers=headers, data=payload)
  collection = []
  soup = BeautifulSoup(response.text,'html.parser')
  for list in soup.find_all('tr'):
    dict = {
      'subject':"",
      'name':"",
      'authority_since':"",
      'WBO': "",
      'duration' : "",
      'address': "",
      'zip': "",
      'city': "",
      'address_title':'',
      'clinic_type':"",
      'import_date':datetime.datetime.now().strftime("%Y-%m-%d %H-%M-%S"),
      'import_link':url
    }
    for idx ,x in  enumerate(list.find_all('td'),start = 0):
      if idx == 0 :
        dict['name'] = x.text
      if idx == 1 :
        address_string = str(x).replace('<td>','').replace('</td>','').split('<br/>')
        dict['zip'] = address_string[len(address_string)-1].split()[0]
        dict['address'] = address_string[1]
        dict['address_title'] = address_string[0]
      if idx == 2 :
        dict['clinic_type'] : x.text
      if idx == 3 :
        dict['duration'] = x.text + ' Monate'
      if idx ==4 :
        dict['subject'] = x.text
    print(dict)
      
    print("==================================================")
    collection.append(dict)
  return collection
```

```python
f = open("collected.json", "a")
collected = []
while current <= limit:
  collected.append(get_data(current))
  current+=1
f.write(json.dumps(collected,indent=2))
f.close()
```
